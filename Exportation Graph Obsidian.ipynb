```python
# Export all the files in an Obsidian in .MD to make a PDF
import re 
import os.path 
import os

# File of the main root of the obsidian
file_source = "YOUR_FILE"

# The name of the file at this end

file_end = 'FICHIER.md'

# Verify if the ligne match with this 
get_file = r"\[([^\]]*)]"

# Recursive function that call itself when she encountered and new file
def Get_Node_Below(basefile):
    # Opening the file we have to check to get the node below
    with open(basefile, mode="r", encoding="utf8") as file_to_open:
        # Opening the file to write in append mode
        with open(file_end, mode="a+", encoding="utf8") as file_to_write:  
            
            # Verify for each line of the file we have to check if there is a node below
            for line in file_to_open:

                # Verify if the file to write is open
                if file_to_write.closed: 
                    file_to_write = open(file_end, mode="a+", encoding="utf8")
                    
                # Verify if the line is a file or a lign
                if not(file_to_write.closed): 
                    # If it's a file check the node below
                    if(line.startswith("[" or line.startswith("[["))):
                        try:
                            match = re.search(get_file, line)
                            line = match.group(1)
                            line = "".join(x for x in line if x not in "[]")
                            line = "".join(x for x in line if x not in "[[]]")
                            file_to_write.close()
                            newbasefile = SearchFile(line)
                            if(newbasefile != basefile):
                                Get_Node_Below(newbasefile) 
                        except:
                            "The file doesn't exist"

                    # If not a file write the lign  
                    else:
                        file_to_write.write(line)

# Check if the file exist in the folder or the folder below
# return the path of the file
def SearchFile(searching_file): 
    for root, directories, files in os.walk("."):  
        for file_name in files:
            if((searching_file + ".md") == file_name):
                 return (os.path.join(root,file_name))

Get_Node_Below(file_source)

```
